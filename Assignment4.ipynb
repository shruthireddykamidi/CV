```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
imgLeft = cv2.imread("/Users/udithracha/Desktop/shruthi/Assignment 3/image1.jpeg", 0)
imgRight = cv2.imread("/Users/udithracha/Desktop/shruthi/Assignment 3/image2.jpeg", 0)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(imgLeft, 'gray')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(imgRight, 'gray')
plt.axis('off')
plt.show()
```

```python
def ShowDisparity(bSize=5):
    # Initialize the stereo block matching object 
    stereo = cv2.StereoBM_create(numDisparities=32, blockSize=bSize)

    # Compute the disparity image
    disparity = stereo.compute(imgLeft, imgRight)

    # Normalize the image for representation
    min = disparity.min()
    max = disparity.max()
    disparity = np.uint8(255 * (disparity - min) / (max - min))
    
    # Plot the result
    return disparity
```

```python
result = ShowDisparity(bSize=5)
plt.imshow(result, 'gray')
plt.axis('off')
plt.show()
```

```python
result = ShowDisparity(bSize=25)
plt.imshow(result, 'gray')
plt.axis('off')
plt.show()
```
